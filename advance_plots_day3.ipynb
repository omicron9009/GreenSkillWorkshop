# Advanced Plots 


```python
# pair plot 
# what do you observed - all the features - looking for relation wrt other - brief idea between any two features 
# diagonals represent higher colinearity 
# others are the scatterplots -
import seaborn as sns 
import pandas as pd 
import matplotlib.pyplot as plt 
data = {
    "Solar":[1200,1600,1300],
    "Wind":[3400,3600,3200],
    "Hydropower":[2900,3100,2800],
    "Biomass": [2500,2700,2400]

}
df=pd.DataFrame(data)
sns.pairplot(df)
plt.show()
```

```python
# heat map - correlation matrix pe plot hota hai 

correlation_matrix=df.corr()
sns.heatmap(correlation_matrix,annot=True,cmap='coolwarm',linewidths=0.5)

plt.title('correlation between energry sources')
plt.show()
```

```python
# box plot 
sns.boxplot(data=df)
plt.title('Distribution ')
plt.show()
```

```python
# violine plot 
# 1 contineous feature and other is categorical feature - used 
sns.violinplot(data=df)
plt.title('violine plot of energy ')
plt.show()
```

```python
energy_values=[100,200,300,400,500]
carbon_emmissions=[10,20,25,40,50]
df_reg=pd.DataFrame(
    {
        "Energy Consuption":energy_values,
        "Carbon Emissions":carbon_emmissions
    }
)

sns.regplot(x='Energy Consuption',y='Carbon Emissions',data=df_reg)
plt.title("Regression plot")
plt.show()
```

# 3d Plot

```python
import numpy as np 
from mpl_toolkits import mplot3d
fig=plt.figure()
ax=plt.axes(projection='3d')

z=np.linspace(-100,100,1000)
x=z*np.sin(5*z)
y=z*np.cos(5*z)

ax.plot3D(x,y,z,'green')
ax.set_title('3D line plot')
plt.show()
```
